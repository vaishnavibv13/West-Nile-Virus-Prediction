```python
def distance(lat1,long1,lat2,long2):
    R=6371
    lat1 = radians(lat1)
    long1 = radians(long1)
    lat2 = radians(lat2)
    long2 = radians(long2)
    dlon = long2 - long1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2 # tangent angle
    c = 2 * atan2(sqrt(a), sqrt(1 - a)) # coordinate
    distance = R * c
    #print(distance)
    return distance
```

```python
breadth=distance(-87.80,41.70,-87.80,41.80)
length=distance(-87.70,41.70,-87.80,41.70)
#print(length,breadth)
area=breadth*length
area=area*247.105#area is in km sq which has to be converted into acre
spray=area*6# 6 boxes are sprayed ,
gallons=1.5*spray*0.0078125#( per acre needs 1.5 fluid ounce of pesticide per acre,converting fluid ounce to gallons)
total_cost=gallons*83.40 # cost per gallon
print("Total cost of Mosquito Control Spray is  "+str(round(total_cost,2))+" for "+str(round(spray,2))+" acres ")
#Non spray locations
breadth=distance(-87.80,41.70,-87.80,41.65)
length=distance(-87.85,41.70,-87.80,41.70)
#print(length,breadth)
area_ns=breadth*length
area_ns=area_ns*247.105#area is in km sq which has to be converted into acre
spray_ns=area_ns*14# 6 boxes are sprayed ,
gallons=1.5*spray_ns*0.0078125#( per acre needs 1.5 fluid ounce of pesticide,converting fluid ounce to gallons)
total_cost_ns=gallons*83.40
print("Total cost of Mosquito Control Spray in unwanted locations is "+str(round(total_cost_ns,2))+" for "+str(round(spray_ns,2))+" acres ")
```

The above graph shows the various features and their correlation with West Nile virus. Based on correlation, the number of mosquitoes is 1 times as likely to detect virus. Maybe there are other carriers of West Nile virus such as birds which do not get affected due to the spray done on land.

The dataset provided gives information on different set of locations (in terms of latitude and longitude) for Spray and trap. This makes it difficult to identify the mosquito trap where virus is detected was sprayed. More information on county-wise spray and trap locations might be helpful in analysing the relationship between spray and virus detection.
